```python
## read in csv
df = pd.read_csv('./leon-events.csv')
```

## Data Cleaning and Processing

```python
## fix column types
df['crash_year'] = df['crash_year'].astype(int)

## fix cases
df['on_street_road_highway'] = df['on_street_road_highway'].str.title()
```

### Crashes on Thomasville Road (01/2011 - 08/2021)

```python
df.query(' on_street_road_highway.str.contains("Thomasville").values ').sort_values(ascending=True, by="crash_year")
```

```python
## save filtered data under new variable, thomasville_df
thomasville_df = df.query(' on_street_road_highway.str.startswith("Thomasville").values').sort_values(ascending=True, by="crash_year")
```
