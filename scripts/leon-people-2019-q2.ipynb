# Import Data

```python
# Import the leon people 2019 Q2 csv with pandas
df = pd.read_csv('../crash-data/quarterly-tranches/leon-people-2019-q2.csv')
df.head(3)
```


What do we want our first map to look like

        Accidents

What do we want to keep

* report_number

* injury_severity
* crash_date_time
* latitude
* longitude
* non_motorist_description_code

```python
# Filter the dataframe to only include the columns we want
crashes = df[["report_number", "crash_year", "role", "injury_severity", "non_motorist_description_code", "crash_date_time", "latitude", "longitude" ]]
crashes.head(3)
```

```python
# sort by report number
crashes_sorted = crashes.sort_values(by=['report_number']).reset_index(drop=True)
crashes_sorted.head(3)
```

Export to CSV

```python
# export to csv
crashes_sorted.to_csv('../crash-data/quarterly-tranches/processed/leon-people-2019-q2.csv', index=False)
```
